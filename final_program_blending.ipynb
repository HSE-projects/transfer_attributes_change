```python
third_table = pd.read_csv('C:/Users/TArs/Desktop/проект Akima/transfer_attributes_change-main/data replaced/Sanya prediction.csv')
third_table = third_table.rename(columns={'Unnamed: 0': 'position'})
mapping = {'Centerbacks':'CB','Midfielders':'MID','Wingers':'WIN', 'Forwards' : 'ST', 'Fullbacks' : 'FB'}
third_table['position'] = third_table['position'].map(mapping)
third_table
```

```python
path_to_Kolya_folder = "C:/Users/TArs/Desktop/проект Akima/transfer_attributes_change-main/data replaced/first_prediction"
path_to_Sanya_folder = "C:/Users/TArs/Desktop/проект Akima/transfer_attributes_change-main/data replaced/second_prediction"
mid_featch = ['sca_per90_y', 'gca_per90_y', 'npxg_xa_per90_y','passes_progressive_distance_y',
'passes_into_penalty_area_y', 'passes_completed_y']
fwd_featch = [ 'sca_per90_y', 'gca_per90_y', 'npxg_per90_y','shots_total_per90_y']
win_featch = [ 'sca_per90_y', 'gca_per90_y', 'dribbles_completed_pct_y','crosses_y','carry_progressive_distance_y']
cb_featch = ['aerials_won_y', 'tackles_won_y','interceptions_y']
fb_featch = ['tackles_won_y', 'crosses_y', 'interceptions_y']
#all_featch = [ 'sca_per90_y', 'gca_per90_y', 'npxg_xa_per90_y','passes_progressive_distance_y',
#'passes_into_penalty_area_y', 'passes_completed_y', 'npxg_per90_y','shots_total_per90_y', 'dribbles_completed_pct_y','crosses_y','carry_progressive_distance_y', 
#'aerials_won_y', 'tackles_won_y','interceptions_y']
player_position = ['MID', 'ST', 'WIN', 'CB', 'FB']
coefficients_val = {}


def average_error(featch, pos, df):
    temp = []
    for i in featch:
            temp.extend(list(df[df['position'] == pos][i]))
    average_error = np.array(temp).mean()
    return average_error

def final_prediction(coefs, pos):
    temp = coefs[pos]
    if pos == 'CB':
        df2 = pd.read_csv(path_to_Kolya_folder + '/CB.csv')
        df3 = pd.read_csv(path_to_Sanya_folder + '/CB.csv')
    elif pos == 'MID':
        df2 = pd.read_csv(path_to_Kolya_folder + '/MID.csv')
        df3 = pd.read_csv(path_to_Sanya_folder + '/MID.csv')
    elif pos == 'ST':
        df2 = pd.read_csv(path_to_Kolya_folder + '/ST.csv')
        df3 = pd.read_csv(path_to_Sanya_folder + '/ST.csv')
    elif pos == 'WIN':
        df2 = pd.read_csv(path_to_Kolya_folder + '/WIN.csv')
        df3 = pd.read_csv(path_to_Sanya_folder + '/WIN.csv')
    elif pos == 'FB':
        df2 = pd.read_csv(path_to_Kolya_folder + '/FB.csv')
        df3 = pd.read_csv(path_to_Sanya_folder + '/FB.csv')
    df2 *= temp[0]
    df3 *= temp[1]
    res = df2 + df3
    return res

def average_model(number):
    temp = {}
    if number == 2:
        table = second_table
    elif number == 3:
        table = third_table
    for i in player_position:
        if i == 'CB':
            featches = cb_featch
        elif i == 'FB':
            featches = fb_featch
        elif i == 'ST':
            featches = fwd_featch
        elif i == 'WIN':
            featches = win_featch
        elif i == 'MID':
            featches = mid_featch
        temp[i] = 1/average_error(featches, i, table)
    return temp    
```

```python
#third_average_mid_val = average_error(mid_featch, 'MID', third_table)
#third_average_fwd_val = average_error(fwd_featch, 'ST', third_table)
#third_average_win_val = average_error(win_featch, 'WIN', third_table)
#third_average_cb_val = average_error(cb_featch, 'CB', third_table)
#third_average_fb_val = average_error(fb_featch, 'FB', third_table)
#third_model_val = {"MID": 1/third_average_mid_val, "ST": 1/third_average_fwd_val, "WIN": 1/third_average_win_val, "CB": 1/third_average_cb_val, "FB": 1/third_average_fb_val}
third_model_val = average_model(3)
third_model_val
```

Coefs for each model

```python
for i in player_position:
        coefficients_val[i] = []
        sum = second_model_val[i] + third_model_val[i]
        coefficients_val[i].append(second_model_val[i] / sum)
        coefficients_val[i].append(third_model_val[i] / sum)
        print(i + ':')
        print('first_model_coef = ' + str(coefficients_val[i][0]))
        print('second_model_coef = ' + str(coefficients_val[i][1]))
        print()
```
